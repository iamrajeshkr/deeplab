```python
# C:\Users\20368750\OneDrive - Larsen & Toubro\Documents\Digital Initiative\SEO\PHASE III_samples\PHASE III\PHASE III\ARCH\B1

import os
import pdfplumber
from docx import Document

def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        return "\n".join([page.extract_text() for page in pdf.pages if page.extract_text()])

def extract_text_from_docx(docx_path):
    doc = Document(docx_path)
    return "\n".join([para.text for para in doc.paragraphs])

# Load all documents in a folder
def load_documents(folder_path):
    text_chunks = []
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if file.endswith(".pdf"):
            text_chunks.append(extract_text_from_pdf(file_path))
        elif file.endswith(".docx"):
            text_chunks.append(extract_text_from_docx(file_path))
    return text_chunks

documents = load_documents(r"C:\Users\20368750\OneDrive - Larsen & Toubro\Documents\Digital Initiative\SEO\PHASE III_samples\PHASE III\PHASE III\ARCH\B1")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_text("\n".join(documents))

# Example: Print the first 3 chunks
print(chunks[:3])
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")  # Efficient embedding model

# Generate embeddings for each chunk
embeddings = np.array([embedding_model.encode(chunk) for chunk in chunks])

# Create FAISS index for retrieval
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)

# Save FAISS index
faiss.write_index(index, "faiss_index.bin")

# Save chunks for reference
import json
with open("chunks.json", "w") as f:
    json.dump(chunks, f)
```

```python
def retrieve_relevant_chunks(query, top_k=1):
    query_embedding = embedding_model.encode(query).reshape(1, -1)
    distances, indices = index.search(query_embedding, top_k)

    # Load chunks
    with open("chunks.json") as f:
        stored_chunks = json.load(f)

    return [stored_chunks[idx] for idx in indices[0]]

# Example query
query = "Is is architectural or structural drawing?"
relevant_chunks = retrieve_relevant_chunks(query)

print("\n".join(relevant_chunks))  # Top 3 relevant chunks
```
